```python
import pandas as pd
import requests
from  bs4 import BeautifulSoup
import time
import wikipediaapi
wiki_wiki = wikipediaapi.Wikipedia('en')


df = pd.read_csv('/content/drive/MyDrive/IndicWiki-OlympicGames/Athletes-Data/Tokyo-2021/Athletes-2021_With_Attributes_Translated.csv')
ImageUrls = []
for ind in df.index:
  print(ind)
  if df['Wiki_Link'][ind] == 'None' or df['Wiki_Link'][ind] == 'AOL':
    continue
  try:
    response=requests.get(df['Wiki_Link'][ind])
    soup = BeautifulSoup(response.text, 'html.parser')
    td_tag = soup.findAll('td', attrs={'class':'infobox-image'})
    img_tag = td_tag[0].findAll('img')
    ImageUrls.append("https:" + img_tag[0]['src'])
  except:
    ImageUrls.append("None")
  if (ind + 1) % 500 == 0:
    LinksDF = pd.DataFrame(ImageUrls)
    LinksDF.to_csv('/content/drive/MyDrive/IndicWiki-OlympicGames/Athletes-Data/Tokyo-2021/Athletes-2021_Image_Links.csv')
LinksDF = pd.DataFrame(ImageUrls)
LinksDF.to_csv('/content/drive/MyDrive/IndicWiki-OlympicGames/Athletes-Data/Tokyo-2021/Athletes-2021_Image_Links.csv')
```
