```python
import pandas as pd
from googlesearch import search
import time
```

```python
DF = pd.read_csv('/Athletes.csv')
DF.head()
```

```python
AthleteName = DF['Name']
AthleteCountry = DF['Country']
urls = []
for i in range(len(AthleteName))[:10000]:
    if AthleteCountry[i] == "None":
        query = AthleteName[i] + " city site:wikipedia.org"
    else:
        query = AthleteName[i] + " " + AthleteCountry[i] + "site:wikipedia.org"
    l1 = []
    print(i + 1, AthleteName[i])
    for j in search(query, num=1, stop=1, pause=20):
        l1.append(j)
    if len(l1) == 0:
        urls.append("None")
    else:
        urls.append(l1[0])
    time.sleep(2)
print(len(urls))
Links = pd.DataFrame(urls)
Links.to_csv('./AthleteLinks1.csv')
# print(urls)
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import time

```

```python
DF = pd.read_csv('/content/Athletes.csv')
print(DF.head())
AthleteName = DF['Name']
AthleteCountry = DF['Country']
urls = []
```

```python
for i in range(len(AthleteName))[:1000]:
    if AthleteCountry[i] == "None":
        query = AthleteName[i] + " site:en.wikipedia.org"
    else:
        query = AthleteName[i] + " " + AthleteCountry[i] + " site:en.wikipedia.org"
    l1 = []
    print(i + 1, AthleteName[i])
    response = requests.get("https://in.search.yahoo.com/search;?p=" + query + "&fr=sfp&iscqry=&fr2=sb-top-search")
    soupObject = BeautifulSoup(response.text, 'html.parser')
    a_tag = soupObject.find_all('a', attrs={'class': ['d-ib', 'ls-05', 'fz-20', 'lh-26', 'td-hu', 'tc', 'va-bot', 'mxw-100p']})
    check = 0
    for a in a_tag:
        if 'en.wikipedia.org' in a['href']:
            urls.append(a['href'])
            print(a['href'])
            check = 1
            break
    if not check:
        urls.append('None')
        print('None')
    Links = pd.DataFrame(urls)
    Links.to_csv('./AthleteLinks1.csv')
print(len(urls))
for u in urls:
    print(u)
```
